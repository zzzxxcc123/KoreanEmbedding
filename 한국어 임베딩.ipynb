```python
# 파일 경로 설정
file_path = '/content/stopwords_kr.txt'

# 파일 읽기
with open(file_path, 'r', encoding='utf-8') as file:
    stopwords = file.read().splitlines()
```

```python
# 불용어 제거
def delete_stopwords(contexts):
  preprocessed_text = []

  for text in contexts:
    text = [w for w in text.split(' ') if w not in stopwords]
    preprocessed_text.append(' '.join(text))

  return preprocessed_text

preprocessed_context = delete_stopwords(contexts)
preprocessed_context
```

```python
# 특수 문자 제거
def remove_special_characters(contexts):
    preprocessed_text = []
    special_characters_set = set('!@#$%^&*(),.')

    for text in contexts:
        text = ''.join([char for char in text if char not in special_characters_set])
        preprocessed_text.append(text)

    return preprocessed_text

preprocessed_context = remove_special_characters(contexts)
preprocessed_context
```

```python
# 반복 횟수가 많음 문자 정규화
from  collections import OrderedDict

def duplicated_sentence_normalizer(context):
  context = list(OrderedDict.fromkeys(context))
  return(context)

preprocessed_context = duplicated_sentence_normalizer(preprocessed_context)
for i, text in enumerate(preprocessed_context):
    print(i, text)
```

```python
# 문장 길이 기반 필터링
# 10 ~ 100
def min_max_filter(min_len, max_len, context):
  preprocessed_text = []

  for text in context:
    if len(text) > min_len and len(text) < max_len:
      preprocessed_text.append(text)

  return preprocessed_text

preprocessed_context = min_max_filter(10, 100, preprocessed_context)
preprocessed_context
```

```python
len(preprocessed_context)
```

```python
# 띄어쓰기 기준으로 단어 분리
cleaned_words_list = []

for context in preprocessed_context:
    words = context.split()
    new_list = []
    for word in words:
      if(len(word) > 1):
        new_list.append(word)
    cleaned_words_list.append(new_list)

print(len(cleaned_words_list))
```

```python
# 토큰화
from tensorflow.keras.preprocessing.text import Tokenizer

tokenizer = Tokenizer()

tokenizer.fit_on_texts(cleaned_words_list)

tokenizer
```

```python
# 단어 사전 만들기
word2idx = tokenizer.word_index
idx2word = {value: key for key, value in word2idx.items() if len(key) > 1}
idx2word
```

```python
encoded = tokenizer.texts_to_sequences(cleaned_words_list)
encoded
```

```python
vocab_size = len(word2idx)
vocab_size
```

```python
from tensorflow.keras.preprocessing.sequence import skipgrams

skip_grams = [skipgrams(sample, vocabulary_size = vocab_size, window_size = 10) for sample in encoded[:100]]

pair, labels = skip_grams[0][0], skip_grams[0][1]
print(pair[:3])
print(labels[:3])
```

```python
# 모델 만들기
from tensorflow.keras.models import Sequential, Model
from tensorflow.keras.layers import Embedding, Reshape, Activation, Input, Dot
from tensorflow.keras.utils import plot_model
import numpy as np

embeding_dim = 100

w_inputs = Input(shape = (1,), dtype = 'int32')

word_enbedding = Embedding(vocab_size, embeding_dim)(w_inputs)

c_inputs = Input(shape = (1,), dtype = 'int32')
context_enbedding = Embedding(vocab_size, embeding_dim)(c_inputs)

dot_product = Dot(axes = 2)([word_enbedding, context_enbedding])
dot_product = Reshape((1,), input_shape = (1,1))(dot_product)
output = Activation('sigmoid')(dot_product)

model = Model(inputs = [w_inputs, c_inputs], outputs = output)
model.summary()
```

```python
model.compile(loss = 'binary_crossentropy', optimizer = 'adam')
```

```python
# 학습 시작
for epoch in range(100):
  loss = 0
  for _, elem in enumerate(skip_grams):
    first_elem = np.array(list(zip(*elem[0]))[0], dtype = 'int32')
    second_elem = np.array(list(zip(*elem[0]))[1], dtype = 'int32')
    labels = np.array(elem[1], dtype = 'int32')
    X = [first_elem, second_elem]
    Y = labels
    loss += model.train_on_batch(X, Y)
  print('Epoch: ', epoch + 1, 'Loss: ',loss)
```

```python
import gensim

f = open('vectors.txt', 'w')
f.write('{} {}\n'.format(vocab_size, embeding_dim))
vectors = model.get_weights()[0]

for word ,i in tokenizer.word_index.items():
    f.write('{} {} \n'.format(word, ' '.join(map(str, list(vectors[i-1, :])))))
f.close()
```
